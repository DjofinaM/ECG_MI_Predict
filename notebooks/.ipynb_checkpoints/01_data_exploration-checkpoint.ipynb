```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import warnings
warnings.filterwarnings('ignore')
```

```python
# Chemin des données
DATA_PATH = r"D:\Documents\memoire\realisation\ECG_MI_Predict\data\raw\train"

# Vérifier si le chemin existe
if not os.path.exists(DATA_PATH):
    print(f"ERREUR : Le chemin n'existe pas : {DATA_PATH}")
else:
    print(f"Chemin des données : {DATA_PATH}")

# Lister les classes (catégories)
classes = os.listdir(DATA_PATH)
print("\nClasses trouvées :", classes)
print("Nombre de classes :", len(classes))

# Afficher le nombre d'images par classe
print("\nNombre d'images par classe :")
for cls in classes:
    class_path = os.path.join(DATA_PATH, cls)
    num_images = len(os.listdir(class_path))
    print(f"  {cls} : {num_images} images")

# Calculer le total d'images
total_images = sum([len(os.listdir(os.path.join(DATA_PATH, cls))) for cls in classes])
print(f"\nTotal d'images dans le dataset : {total_images}")
```

```python
# Afficher une image de chaque classe
n_cols = min(5, len(classes))
n_rows = 1
fig, axes = plt.subplots(n_rows, n_cols, figsize=(4*n_cols, 4))

if n_cols == 1:
    axes = [axes]

for idx, cls in enumerate(classes[:n_cols]):
    class_path = os.path.join(DATA_PATH, cls)
    img_name = os.listdir(class_path)[0]  # Première image
    img_path = os.path.join(DATA_PATH, cls, img_name)
    
    # Charger l'image
    img = cv2.imread(img_path)
    if img is not None:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        axes[idx].imshow(img_rgb)
        axes[idx].set_title(f"{cls}\n{img.shape[1]}x{img.shape[0]}")
        axes[idx].axis('off')
    else:
        axes[idx].text(0.5, 0.5, f"Erreur\n{img_name}", 
                      ha='center', va='center', transform=axes[idx].transAxes)
        axes[idx].set_title(f"{cls}\nImage non lisible")
        axes[idx].axis('off')

plt.suptitle("Exemple d'une image par classe", fontsize=16, y=1.05)
plt.tight_layout()
plt.show()
```

```python
print("=" * 70)
print("ANALYSE COMPLÈTE DES TAILLES ET FORMATS D'IMAGES")
print("=" * 70)

# Variables de collecte
image_sizes = []
image_shapes = []
formats_count = {}
problematic_images = []
all_images_info = []

# Compter le nombre total d'images
total_images = 0
for cls in classes:
    class_path = os.path.join(DATA_PATH, cls)
    total_images += len(os.listdir(class_path))

print(f"Analyse de {total_images} images...\n")

# Parcourir toutes les images (sans tqdm)
images_processed = 0
for cls in classes:
    class_path = os.path.join(DATA_PATH, cls)
    
    for img_name in os.listdir(class_path):
        img_path = os.path.join(DATA_PATH, cls, img_name)
        img = cv2.imread(img_path)
        
        if img is None:
            problematic_images.append((cls, img_name, "Impossible de lire"))
            images_processed += 1
            continue
        
        # Vérifier si l'image est vide
        if img.size == 0:
            problematic_images.append((cls, img_name, "Image vide"))
            images_processed += 1
            continue
        
        h, w, c = img.shape
        
        # Collecter les informations
        image_sizes.append((w, h))  # (largeur, hauteur)
        image_shapes.append((h, w, c))
        
        # Format d'image
        _, ext = os.path.splitext(img_name)
        ext = ext.lower()
        formats_count[ext] = formats_count.get(ext, 0) + 1
        
        # Informations détaillées
        all_images_info.append({
            'class': cls,
            'name': img_name,
            'width': w,
            'height': h,
            'channels': c,
            'format': ext,
            'size_kb': os.path.getsize(img_path) / 1024
        })
        
        images_processed += 1
        
        # Afficher la progression toutes les 100 images
        if images_processed % 100 == 0:
            print(f"  Traité {images_processed}/{total_images} images...")

print(f"\nAnalyse terminée ! {images_processed} images traitées.")

# Analyse des tailles
unique_sizes = Counter(image_sizes)

print(f"\n📊 RÉSULTATS DE L'ANALYSE :")
print("-" * 50)
print(f"Images analysées avec succès : {len(image_sizes)}")
print(f"Images problématiques : {len(problematic_images)}")
print(f"Tailles différentes : {len(unique_sizes)}")
print(f"Formats différents : {len(formats_count)}")

# Affichage des tailles
print(f"\n🔍 DISTRIBUTION DES TAILLES :")
if len(unique_sizes) <= 15:
    print("Taille (largeur x hauteur) | Nombre | Pourcentage")
    print("-" * 50)
    for size, count in sorted(unique_sizes.items(), key=lambda x: x[1], reverse=True):
        percentage = (count / len(image_sizes)) * 100
        print(f"{size[0]:4d} x {size[1]:4d} | {count:6d} | {percentage:6.2f}%")
else:
    print("Top 15 des tailles les plus fréquentes :")
    print("Taille (largeur x hauteur) | Nombre | Pourcentage")
    print("-" * 50)
    for size, count in unique_sizes.most_common(15):
        percentage = (count / len(image_sizes)) * 100
        print(f"{size[0]:4d} x {size[1]:4d} | {count:6d} | {percentage:6.2f}%")

# Affichage des formats
print(f"\n📁 FORMATS D'IMAGES TROUVÉS :")
print("Format | Nombre | Pourcentage")
print("-" * 50)
for fmt, count in sorted(formats_count.items(), key=lambda x: x[1], reverse=True):
    percentage = (count / len(image_sizes)) * 100
    print(f"{fmt:6s} | {count:6d} | {percentage:6.2f}%")
```

```python
print("\n" + "=" * 70)
print("STATISTIQUES DÉTAILLÉES DES DIMENSIONS")
print("=" * 70)

# Extraire les dimensions
widths = [w for w, h in image_sizes]
heights = [h for w, h in image_sizes]
channels = [c for h, w, c in image_shapes]

# Calcul des statistiques pour la largeur
width_stats = {
    'min': min(widths),
    'max': max(widths),
    'mean': np.mean(widths),
    'median': np.median(widths),
    'std': np.std(widths)
}

# Calcul des statistiques pour la hauteur
height_stats = {
    'min': min(heights),
    'max': max(heights),
    'mean': np.mean(heights),
    'median': np.median(heights),
    'std': np.std(heights)
}

# Affichage des statistiques
print("\n📏 STATISTIQUES DE LARGEUR :")
print(f"  Minimum  : {width_stats['min']:5d} pixels")
print(f"  Maximum  : {width_stats['max']:5d} pixels")
print(f"  Moyenne  : {width_stats['mean']:7.1f} pixels")
print(f"  Médiane  : {width_stats['median']:7.1f} pixels")
print(f"  Écart-type : {width_stats['std']:7.1f} pixels")

print("\n📐 STATISTIQUES DE HAUTEUR :")
print(f"  Minimum  : {height_stats['min']:5d} pixels")
print(f"  Maximum  : {height_stats['max']:5d} pixels")
print(f"  Moyenne  : {height_stats['mean']:7.1f} pixels")
print(f"  Médiane  : {height_stats['median']:7.1f} pixels")
print(f"  Écart-type : {height_stats['std']:7.1f} pixels")

# Rapport d'aspect
aspect_ratios = [w/h for w, h in zip(widths, heights)]
print("\n📐 RAPPORTS D'ASPECT (Largeur/Hauteur) :")
print(f"  Minimum  : {min(aspect_ratios):.3f}")
print(f"  Maximum  : {max(aspect_ratios):.3f}")
print(f"  Moyenne  : {np.mean(aspect_ratios):.3f}")
print(f"  Médiane  : {np.median(aspect_ratios):.3f}")

# Canaux
print(f"\n🎨 CANAUX DE COULEUR :")
print(f"  Toutes les images ont {set(channels)} canaux")

# Histogramme des dimensions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Histogramme des largeurs
axes[0].hist(widths, bins=30, edgecolor='black', alpha=0.7)
axes[0].axvline(width_stats['mean'], color='red', linestyle='--', label=f'Moyenne: {width_stats["mean"]:.1f}')
axes[0].axvline(width_stats['median'], color='green', linestyle='--', label=f'Médiane: {width_stats["median"]:.1f}')
axes[0].set_xlabel('Largeur (pixels)')
axes[0].set_ylabel('Fréquence')
axes[0].set_title('Distribution des Largeurs')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Histogramme des hauteurs
axes[1].hist(heights, bins=30, edgecolor='black', alpha=0.7, color='orange')
axes[1].axvline(height_stats['mean'], color='red', linestyle='--', label=f'Moyenne: {height_stats["mean"]:.1f}')
axes[1].axvline(height_stats['median'], color='green', linestyle='--', label=f'Médiane: {height_stats["median"]:.1f}')
axes[1].set_xlabel('Hauteur (pixels)')
axes[1].set_ylabel('Fréquence')
axes[1].set_title('Distribution des Hauteurs')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Nuage de points largeur vs hauteur
axes[2].scatter(widths, heights, alpha=0.5, s=10)
axes[2].set_xlabel('Largeur (pixels)')
axes[2].set_ylabel('Hauteur (pixels)')
axes[2].set_title('Relation Largeur-Hauteur')
axes[2].grid(True, alpha=0.3)

# Ligne de rapport 1:1
max_dim = max(max(widths), max(heights))
axes[2].plot([0, max_dim], [0, max_dim], 'r--', alpha=0.5, label='Ratio 1:1')

# Calculer la ligne de régression
if len(widths) > 1:
    z = np.polyfit(widths, heights, 1)
    p = np.poly1d(z)
    axes[2].plot([min(widths), max(widths)], [p(min(widths)), p(max(widths))], 'g-', alpha=0.8, label='Tendance')

axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
print("\n" + "=" * 70)
print("ANALYSE DE LA QUALITÉ DES TRACÉS ECG")
print("=" * 70)

def analyze_image_quality(img):
    """
    Analyse la qualité d'une image pour déterminer si le tracé ECG est propre ou bruité
    """
    # Convertir en niveaux de gris
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    # 1. Mesure du contraste (écart-type des intensités)
    contrast = np.std(gray)
    
    # 2. Mesure de la netteté (variance du Laplacien)
    laplacian = cv2.Laplacian(gray, cv2.CV_64F)
    sharpness = np.var(laplacian)
    
    # 3. Détection des bords (pour évaluer la définition du tracé)
    edges = cv2.Canny(gray, 50, 150)
    edge_density = np.sum(edges > 0) / (gray.shape[0] * gray.shape[1])
    
    # 4. Mesure du bruit (rapport signal/bruit estimé)
    # Approximation: plus l'écart-type est élevé par rapport à la moyenne, plus il y a de variations
    mean_intensity = np.mean(gray)
    noise_ratio = contrast / mean_intensity if mean_intensity > 0 else 0
    
    # 5. Uniformité du fond
    # On considère que le fond doit être relativement uniforme pour un ECG propre
    # Calculer l'histogramme et vérifier les pics
    hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
    hist = hist.flatten()
    hist_peaks = np.argsort(hist)[-3:]  # Les 3 intensités les plus fréquentes
    background_uniformity = hist[hist_peaks[-1]] / np.sum(hist)  # Proportion du pic principal
    
    # Classification
    if contrast > 60 and sharpness > 100 and background_uniformity > 0.3:
        quality = "Propre"
    elif contrast < 30 or sharpness < 20:
        quality = "Bruité"
    else:
        quality = "Moyenne"
    
    return {
        'contrast': contrast,
        'sharpness': sharpness,
        'edge_density': edge_density,
        'noise_ratio': noise_ratio,
        'background_uniformity': background_uniformity,
        'quality': quality
    }

# Analyser un échantillon d'images
print("Analyse de la qualité sur un échantillon d'images...")

# Préparer l'affichage
n_classes_to_show = min(4, len(classes))
n_images_per_class = 3

fig, axes = plt.subplots(n_images_per_class, n_classes_to_show, 
                         figsize=(4*n_classes_to_show, 3.5*n_images_per_class))

if n_classes_to_show == 1:
    axes = axes.reshape(n_images_per_class, 1)

# Collecter les métriques de qualité
all_quality_metrics = []

for col_idx, cls in enumerate(classes[:n_classes_to_show]):
    class_path = os.path.join(DATA_PATH, cls)
    img_names = os.listdir(class_path)[:n_images_per_class]
    
    for row_idx, img_name in enumerate(img_names):
        img_path = os.path.join(DATA_PATH, cls, img_name)
        img = cv2.imread(img_path)
        
        if img is None:
            continue
            
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Analyser la qualité
        quality_metrics = analyze_image_quality(img_rgb)
        quality_metrics['class'] = cls
        quality_metrics['image_name'] = img_name
        all_quality_metrics.append(quality_metrics)
        
        # Afficher l'image
        ax = axes[row_idx, col_idx]
        ax.imshow(img_rgb)
        
        # Ajouter les informations de qualité
        quality = quality_metrics['quality']
        color = 'green' if quality == "Propre" else 'red' if quality == "Bruité" else 'orange'
        
        title = f"{cls}\n"
        title += f"Contraste: {quality_metrics['contrast']:.1f}\n"
        title += f"Qualité: {quality}"
        
        ax.set_title(title, fontsize=9, color=color)
        ax.axis('off')
        
        # Ajouter des infos de taille
        h, w = img.shape[:2]
        ax.text(0.02, 0.98, f"{w}x{h}", 
                transform=ax.transAxes, fontsize=7,
                verticalalignment='top', color='white',
                bbox=dict(boxstyle='round', facecolor='black', alpha=0.5))

plt.suptitle("Analyse de la qualité des tracés ECG", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()

# Statistiques de qualité
print("\n📈 STATISTIQUES DE QUALITÉ :")
print("-" * 50)

if all_quality_metrics:
    qualities = [m['quality'] for m in all_quality_metrics]
    contrasts = [m['contrast'] for m in all_quality_metrics]
    sharpnesses = [m['sharpness'] for m in all_quality_metrics]
    
    quality_counts = Counter(qualities)
    
    print("Distribution de la qualité :")
    for quality, count in quality_counts.items():
        percentage = (count / len(all_quality_metrics)) * 100
        print(f"  {quality}: {count} images ({percentage:.1f}%)")
    
    print(f"\nContraste moyen: {np.mean(contrasts):.1f} ± {np.std(contrasts):.1f}")
    print(f"Netteté moyenne: {np.mean(sharpnesses):.1f} ± {np.std(sharpnesses):.1f}")
    
    print("\n🔍 INTERPRÉTATION :")
    print("-" * 50)
    print("ÉCHELLE DE QUALITÉ :")
    print("  Propre   : Contraste > 60, Netteté > 100, Fond uniforme")
    print("  Moyenne  : Entre les critères 'Propre' et 'Bruité'")
    print("  Bruité   : Contraste < 30 ou Netteté < 20")
    
    print("\nCARACTÉRISTIQUES DES ECG PROPRES :")
    print("  ✓ Tracés nets et bien définis")
    print("  ✓ Fond uniforme (noir ou blanc)")
    print("  ✓ Bon contraste entre le tracé et le fond")
    print("  ✓ Peu d'artefacts ou de bruit")
    
    print("\nCARACTÉRISTIQUES DES ECG BRUITÉS :")
    print("  ✗ Tracés flous ou granuleux")
    print("  ✗ Fond non uniforme ou taché")
    print("  ✗ Faible contraste")
    print("  ✗ Présence d'artefacts, lignes parasites")
else:
    print("Aucune métrique de qualité disponible.")
```

```python
# Fonctions utiles pour le prétraitement
def resize_image(img, target_size=(224, 224)):
    """Redimensionne une image à la taille cible"""
    return cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)

def normalize_image(img):
    """Normalise les intensités de l'image entre 0 et 1"""
    return img / 255.0

def enhance_contrast(img):
    """Améliore le contraste d'une image"""
    # Conversion en niveaux de gris
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Égalisation d'histogramme
    enhanced = cv2.equalizeHist(gray)
    # Retour en RGB
    return cv2.cvtColor(enhanced, cv2.COLOR_GRAY2RGB)

def denoise_image(img):
    """Réduit le bruit d'une image"""
    return cv2.medianBlur(img, 3)

def preprocess_ecg_image(img_path, target_size=(224, 224)):
    """
    Pipeline complet de prétraitement pour une image ECG
    """
    # Charger l'image
    img = cv2.imread(img_path)
    if img is None:
        return None
    
    # Convertir en RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Redimensionner
    img = resize_image(img, target_size)
    
    # Réduire le bruit
    img = denoise_image(img)
    
    # Améliorer le contraste
    img = enhance_contrast(img)
    
    # Normaliser
    img = normalize_image(img)
    
    return img

print("Fonctions de prétraitement définies avec succès.")
print("Utilisez preprocess_ecg_image() pour prétraiter une image individuelle.")
```
